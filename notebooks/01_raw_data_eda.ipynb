# Raw Data Exploratory Data Analysis

This notebook inspects the raw GEO files before data wrangling.

The aim is to understand the structure of the expression matrix and metadata file, verify how samples are represented, and confirm that sample descriptions from the series matrix match the sample columns in the expression matrix.

```python
#IMPORTS
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#SET FILE PATHS
series_file = "../data/raw_data/GSE147379_series_matrix.txt"
expression_file = "../data/raw_data/GSE147379_Corton_29_Chem_data_matrix.txt"
```

```python
#LOAD EXPRESSION MATRIX & INSPECT STRUCTURE
expression = pd.read_csv(expression_file, sep="\t")
expression.head()
```

```python
#READ SERIES METADATA FILE
with open(series_file, errors="ignore") as f:
    reader = csv.reader(f, delimiter="\t")
    rows = list(reader)
```

```python
#EXTRACT SAMPLE METADATA- CHEMICAL, DOSE, SAMPLE NAME (LATER USED TO FIND EXPRESSION VALS)
sample_desc = None
sample_chem_row = None
sample_dose_row = None

for row in rows:
    if len(row) > 0 and row[0] == "!Sample_description":
        sample_desc = row[1:]
        break

for row in rows:
    if len(row) > 1 and row[0] == "!Sample_characteristics_ch1":
        if row[1].lower().startswith("chemical:"):
            sample_chem_row = row[1:]
            break

for row in rows:
    if len(row) > 1 and row[0] == "!Sample_characteristics_ch1":
        if row[1].lower().startswith("dose:"):
            sample_dose_row = row[1:]
            break
```

The sample description row was extracted because it contains the sample names used as column headers in the expression matrix. The chemical and dose rows were extracted to identify the exposure condition for each sample.

```python
#INSPECT METADATA VALUES
print("Sample descriptions:")
print(sample_desc[:5])

print("\nChemical row:")
print(sample_chem_row[:5])

print("\nDose row:")
print(sample_dose_row[:5])
```

```python
#MAKE A DATAFRAME (TO INDEX INTO CHEMICAL OF INTEREST)
metadata_preview = pd.DataFrame({
    "sample_column": sample_desc,
    "chemical": sample_chem_row,
    "dose": sample_dose_row
})

metadata_preview.head()
```

```python
#METADATA FOR SAMPLES WITH CHEMICAL OF INTEREST- S-BIOALLETHRIN
target_chemical = "S-bioallethrin"

s_bio_metadata = metadata_preview[
    metadata_preview["chemical"].str.contains(target_chemical, case=False, na=False)
].copy()

s_bio_metadata
```

```python
#GET SAMPLE NAMES FOR CHEMICAL OF INTEREST
s_bio_sample_names = list(s_bio_metadata["sample_column"])

s_bio_sample_names
```

```python
expression_sample_columns = list(expression.columns[1:])

matching_sbio_samples = [
    sample for sample in s_bio_sample_names
    if sample in expression_sample_columns
]

missing_sbio_samples = [
    sample for sample in s_bio_sample_names
    if sample not in expression_sample_columns
]

print("S-bioallethrin samples in metadata:", len(s_bio_sample_names))
print("Matching samples in expression matrix:", len(matching_sbio_samples))
print("Missing samples:", len(missing_sbio_samples))

print("\nMissing sample names:")
print(missing_sbio_samples)
```

ALL SAPLES WITH VARYING DOSES OF OUR TARGET CHEMICAL (S-BIOALLETHRIN) WERE IDENTIFIED IN THE SERIES MATRIX METADATA FILE.
BASED ON THE SAMPLE DESCRIPTION (SAMPLE NAMES),THE PRESENCE OF THOSE SAMPLES WAS ALSO IDENTIFIED IN THE EXPRESSION MATRIX FILE.
